## Importing all necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Creating custom range (row) and velocity (column) arrays

```python
row_values = np.arange(5,115,5)
column_values = np.arange(500,1175,25)
```

## Using a nested for-loop to populate an empty list with values calculating projectile time to target for each custom range and velocity

```python
data = []

for row in row_values:
    row_data =[]
    
    for col in column_values:
        row_data.append(row / col)
        
    data.append(row_data)
```

## Createing a dataframe from the list values with modified column and row names, showing projectile time to target in seconds


```python
modified_column_names= [str(col) +'(m/s)' for col in column_values]
modified_row_names =[str(row) +'(m)' for row in row_values]

df = pd.DataFrame(data, index = modified_row_names, columns = modified_column_names)

display(df)
```

# Creating heatmap for projectile time to target

```python
Bullet_Travel_Time = df.round(3)

plt.subplots(dpi=100, figsize=(20, 15))
sns.heatmap(Bullet_Travel_Time , annot=True, cbar=True,  xticklabels=column_values, yticklabels=row_values)

plt.title('Projectile Time to Target (seconds)')
plt.xlabel('Velocity (meters/second)')
plt.ylabel('Distance To Target (meters)')
# Placing the column names on top of the heatmap
plt.tick_params(axis='x', which='both', bottom=True, top=True, labeltop=True, labelbottom=True)
```

# Introducing parameters

```python
Tac_Sprint_Speed = 7.5 #meters/second
Player_width = 0.6 #meters
```

# Creating a new dataframe to determine the distance a player can travel after being fired at while they are sprinting  

```python
df2 = pd.DataFrame(Bullet_Travel_Time.values*Tac_Sprint_Speed, index=Bullet_Travel_Time.index, columns=Bullet_Travel_Time.columns)

Distance_Player_Traveled_Post_Shot = df2.round(4)

display(df2)
```

# Creating a graph to show relationship between velocity and time to target impact

```python
Ploting_single_row= Bullet_Travel_Time.loc['110(m)'].values
# Get the column values as a list for the x-axis



plt.figure(dpi=100, figsize=(8, 6))
plt.plot(column_values, Ploting_single_row, marker='o')  # You can customize the marker or linestyle if needed
plt.title('Projectile Travel Time (110 meters)')
plt.xlabel('Velocity (meters/second)')
plt.ylabel('Time to Impact (seconds)')
plt.grid(True)

```

## Creating a new dataframe to determine if player travel distance exceeds player width post shot

```python
exceeded_player_width = Distance_Player_Traveled_Post_Shot > Player_width

# Create a new DataFrame showing whether the player width is exceeded post-shot
result_df = pd.DataFrame(exceeded_player_width, index=Distance_Player_Traveled_Post_Shot.index, columns=Distance_Player_Traveled_Post_Shot.columns)

display(result_df)
```

## Creating a heatmap to showcase if your projectile will hit its intended target at various distances and weapon velocitys while simply aiming at the center mass of your target, when that target is sprinting perpendicular to the observers position.

```python
# Set the threshold value
threshold = Player_width  

colors = ['green', 'red']
#custom color map
cmap = sns.color_palette(colors)

# Plotting the heatmap with a diverging colormap based on the threshold
plt.subplots(dpi=100, figsize=(20, 15))
sns.heatmap( Distance_Player_Traveled_Post_Shot, annot=True, cmap= cmap, center=threshold, cbar=False,  xticklabels=column_values, yticklabels=row_values)
plt.title('Success Heatmap: Hitting Moving Targets Without Leading Projectile')
plt.xlabel('Velocity (meters/second)')
plt.ylabel('Distance To Target (meters)')

# Placing the column names on top of the heatmap
plt.tick_params(axis='x', which='both', bottom=True, top=True, labeltop=True, labelbottom=True)
```
